# Calculadora em python

 Ã‰ uma calculadora onde o usuario insere um numero na aba numero e escolhe um sinal na aba "algoritimo" e terminando de inseririr na aba numero2.
  Depois que coloca o numero + sinal + numero2 o codigo vai retornar com o valor



```python

Numero = int(input('numero:'))
Algoritimo = input('algoritimo:')
Numero2 = int(input('numero2 :'))
if Algoritimo == '+':
  soma = Numero+Numero2
  print(soma)
elif Algoritimo == '-':
  subtracao = Numero-Numero2
  print(subtracao)
elif Algoritimo == '*':
  Multiplicacao = Numero*Numero2
  print(Multiplicacao)
elif Algoritimo == '/':
  divisao = Numero/Numero2
  print(divisao)




```
